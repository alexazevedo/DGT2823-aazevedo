# Trabalho Prático - Limpeza de Dados

## Leitura do arquivo CSV

```python
import pandas as pd

dados = pd.read_csv('dados.csv', sep=';', engine='python')
```

## Verificação dos dados importados

```python
print(dados.info())
```

```python
print(dados.head(10))
```

```python
print(dados.tail(10))
```

## Criação de cópia do conjunto de dados

```python
dados_copia = dados.copy()
```

## Substituição de valores nulos em Calories por 0

```python
dados_copia['Calories'] = dados_copia['Calories'].fillna(0)
print(dados_copia)
```

## Substituição de valores nulos em Date por '1900/01/01'

```python
dados_copia['Date'] = dados_copia['Date'].fillna('1900/01/01')
print(dados_copia)
```

## Conversão da coluna Date para string e correção do valor 20201226

```python
dados_copia['Date'] = dados_copia['Date'].astype(str)
dados_copia.loc[dados_copia['Date'] == '20201226', 'Date'] = "'2020/12/26'"
```

## Correção do valor '1900/01/01' para NaN

```python
dados_copia['Date'] = dados_copia['Date'].replace('1900/01/01', pd.NA)
print(dados_copia)
```

## Conversão da coluna Date para datetime

```python
dados_copia['Date'] = pd.to_datetime(dados_copia['Date'], format="'%Y/%m/%d'", errors='coerce')
print(dados_copia)
```

## Remoção de registros com valores nulos em Date

```python
dados_copia = dados_copia.dropna(subset=['Date'])
print(dados_copia)
```
